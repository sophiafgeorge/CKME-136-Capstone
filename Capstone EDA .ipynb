We eill now convert our dataset into a matrix with the movie_titles as the columns, the user_id as the index and the ratings as the values. The columns in the matrix will be  the movie_titles and the rows will be the user_ids. Each column represents all the ratings of a movie by all the users. The rating appear as NAN where a user didn't rate a certain movie. We shall use this matrix to compute the correlation between the ratings of a single movie and the rest of the movies in the matrix. We will use pandas pivot_table function to create the movie matrix.

```python
movie_matrix = df.pivot_table(index='user_id', columns='movie_title', values='rating')
movie_matrix.head()
```

```python
movie_matrix.describe()

# get more statistical information for each movie, such has mean, std, max, min and quartiles
```

We will now look at  the most rated movies and choose one of them to work with in this simple recommender system. We use pandas sort_values function and set ascending to false in order to sort the movies from the most rated. We then use the head() function to view the top 10.

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
# starwars is the most rated movie
```

Lets assume that a user has watched Star Wars (1977). We would like like to recommend movies to this user based on his watching history. The goal is to look for movies that are similar to Star Wars1977 which we shall recommend to this user. We can achieve this by computing the correlation between these movies ratings and the ratings of the rest of the movies in the dataset. The first step is to create a dataframe with the ratings of Star Wars.

```python
star_wars_rating = movie_matrix['Star Wars (1977)']
star_wars_rating.head()
```

```python
similar_to_star_wars = movie_matrix.corrwith(star_wars_rating)
```

```python
similar_to_star_wars.head()
```

```python
 similar_to_star_wars.sort_values(ascending=False).head()
    # get top 10 movies most corelated with  star wars
```

We realize from the computation that there is a very strong positive correlation (of 1.0) between star Wars (1977) and No Escape (1994)

Our matrix hasmany missing values since not all the movies were rated by all the users. We therefore drop those null values and transform correlation results into dataframes to make the results look more appealing.

```python
corr_star_wars = pd.DataFrame(similar_to_star_wars, columns=['Correlation'])
corr_star_wars.dropna(inplace=True)
corr_star_wars.head()
```

 We have an issue, some of the movies have very few ratings and may end up being recommended because one or two people gave them a 5 star rating. We can fix this by setting a threshold for the number of ratings. From the histogram earlier we saw a sharp decline in number of ratings from 100. We shall therefore set this as the threshold.hn. In order to do this we need to join the star wars dataframe with the number_of_ratings column in the ratings dataframe.

```python
corr_star_wars = corr_star_wars.join(ratings['number_of_ratings'])
```

```python
corr_star_wars.head()
```

 Now obtain the movies that are most similar to Star Wars (1977) by limiting them to movies that have at least 100 reviews. We then sort them by the correlation column and view the first 10.

```python
corr_star_wars[corr_star_wars['number_of_ratings'] > 100].sort_values(by='Correlation', ascending=False).head(10)
# no surprise that the most correlated movies is Star Wars itself,other similar movies invlude 
#Empire Strikes Back, Return of the Jedi and Raiders of the Lost Arc. All of which makes perfect sense.
```
