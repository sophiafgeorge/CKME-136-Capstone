We will now convert our dataset into a matrix with the movie_titles as the columns, the user_id as the index and the ratings as the values. The columns in the matrix will be the movie_titles and the rows will be the user_ids. Each column represents all the ratings of a movie by all the users. The rating appear as NAN where a user didn't rate a certain movie. We shall use this matrix to compute the correlation between the ratings of a single movie and the rest of the movies in the matrix. We will use pandas pivot_table function to create the movie matrix.

```python
movie_matrix = df.pivot_table(index='user_id', columns='movie_title', values='rating')
movie_matrix.head()
```

```python
movie_matrix.describe()

# get more statistical information for each movie, such has mean, std, max, min and quartiles
```
