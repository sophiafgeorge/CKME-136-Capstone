```python
import pandas as pd 
import numpy as np
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
user_info= pd.read_csv('u.data', sep='\t', 
                 names=['user_id','movie_id','rating','timestamp'])
```

```python
user_info.head()
```

```python
movie_info = pd.read_csv('u.item', sep='|',
                         names = ['movie_id', 
                         'movie_title', 'release_date', 
                         'video_release_date','imdb_url','unknown', 
                         'action','adventure','animation', 'children',
                         'comedy', 'crime', 'documentary','drama',
                         'fantasy','film_noir', 'horror', 'musical',
                         'mystery','romance', 'sci_fi', 'thriller',
                         'war','western'], encoding='latin-1')
```

```python
movie_info.head()
```

```python
user_demo_info = pd.read_csv('u.user', sep='|', 
                 names=['user_id','age','gender','occupation',
                        'zipcode'])
```

```python
user_demo_info.head()
```

```python
user = pd.merge(user_demo_info, user_info, on='user_id')
df = pd.merge(user, movie_info, on='movie_id')
pd.set_option('display.max_columns', 500)
df.describe()
```

```python
df
```

```python
df = df.drop(columns='timestamp')
```

```python
df = df.drop(columns=['video_release_date', 'imdb_url'])
```

```python
df = df.drop(columns='release_date')
```

```python
df = df.drop(columns='zipcode')
```

```python
df.describe()
```

```python
df
```

```python
popular_movies = pd.DataFrame(df.groupby('movie_title')['rating'].mean())
popular_movies.head()
```

```python
popular_movies ['number_of_ratings'] = df.groupby('movie_title')['rating'].count()
popular_movies.head()
```

```python
popular_movies = popular_movies [popular_movies['rating']> 4].sort_values(by= 'number_of_ratings', ascending =False).head(10)
```
